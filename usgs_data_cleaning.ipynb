```python
# cleaned_df = rain_df.resample('1H', on="Date_Time").agg({'Sitka_rain_mm_Tot': np.sum, 
#                                                          'Sitka_ec5_vwc_sp2_40cm_mean': np.mean, 
#                                                          'Sitka_ec5_vwc_sp2_20cm_mean': np.mean, 
#                                                          'Sitka_waterPressure_psig_sp2_60cm_mean': np.mean,
#                                                          'Sitka_ec5_vwc_sp2_40cm_max': np.max, 
#                                                          'Sitka_ec5_vwc_sp2_20cm_max': np.max, 
#                                                          'Sitka_waterPressure_psig_sp2_60cm_max': np.max,
#                                                          'Sitka_ec5_vwc_sp2_40cm_dt': np.max,
#                                                          'Sitka_ec5_vwc_sp2_20cm_dt': np.max,
#                                                          'Sitka_waterPressure_psig_sp2_60cm_dt': np.max
#                                                         })

cleaned_df = rain_df.resample('24H', on="Date_Time", origin='start_day').agg({'Sitka_rain_mm_Tot': np.sum, 
                                                         'Sitka_ec5_vwc_sp2_40cm_mean': np.mean, 
                                                         'Sitka_ec5_vwc_sp2_20cm_mean': np.mean, 
                                                         'Sitka_waterPressure_psig_sp2_60cm_mean': np.mean,
                                                         'Sitka_ec5_vwc_sp2_40cm_max': np.max, 
                                                         'Sitka_ec5_vwc_sp2_20cm_max': np.max, 
                                                         'Sitka_waterPressure_psig_sp2_60cm_max': np.max,
                                                         'Sitka_ec5_vwc_sp2_40cm_dt': np.max,
                                                         'Sitka_ec5_vwc_sp2_20cm_dt': np.max,
                                                         'Sitka_waterPressure_psig_sp2_60cm_dt': np.max
                                                        })

cleaned_df.dropna(subset=['Sitka_rain_mm_Tot'], inplace=True)
```
