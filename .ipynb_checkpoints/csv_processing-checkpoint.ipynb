```python
ocr = pd.read_csv('./sitka_data/combined_3_hour_wind.csv', index_col='Date_Time')#,sep='\t')
#ocr = pd.read_csv('./SP1.csv')

# ocr["Date_Time"] = ocr["timestamp_utc"]
# for ii in range(len(ocr["timestamp_utc"])):
#     ocr["Date_Time"][ii] = OrdinalToDatetime(ocr["timestamp_utc"][ii])

#ocr.index = pd.to_datetime(ocr["Date_Time"])
#ocr.drop(["timestamp_utc"], axis=1, inplace=True)

# add in forecasted rainfall, range(start, end, interval)
for i in tqdm_notebook(range(1,72,1)):
    ocr = add_max_rainfall(ocr, i, 0, noise=False)
    
#df.sort_index(axis=1, inplace=True) 

pp_cols = [v for v in ocr.columns if "Pressure_" in v]  # allows any pp depths, as the depth varies from pit to pit
VWC_cols = [v for v in ocr.columns if "vwc_" in v]  # allows any water content depths, as the depth varies from pit to pit
ocr[pp_cols] = ocr[pp_cols].mask(ocr[pp_cols] < 0, 0) #mask pp smaller than -14, we're interested in larger pp
    
cols_at_end = VWC_cols + pp_cols
ocr = ocr[[c for c in ocr if c not in cols_at_end] + [c for c in cols_at_end if c in ocr]]

ocr.to_csv('./sitka_data/processed_for_training_3_hour_72_1_mask_neg_pp.csv')
```
